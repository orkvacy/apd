```python
import pandas as pd

# 1. Membaca dataset
print("===== 1. Membaca Dataset =====")
df = pd.read_csv('gym_membership.csv')
print(f"Jumlah total data: {len(df)} baris\n")







# 5. Menampilkan jumlah data yang kosong
print("===== 5. Jumlah Data Kosong per Kolom =====")
null_counts = df.isnull().sum()
for column, count in null_counts.items():
    print(f"{column}: {count} data kosong")
print("\n")

# 6. Analisis berdasarkan kategori umur
print("===== 6. Analisis Berdasarkan Kategori Umur =====")
# Membuat kategori umur
df['age_category'] = df['Age'].apply(lambda x: '1-18 tahun' if x <= 18 else '18+ tahun')

# Statistik per kategori umur
age_stats = df.groupby('age_category').agg({
    'id': 'count',
    'avg_time_in_gym': ['mean', 'min', 'max']
}).round(2)

print("Statistik per kategori umur:")
for category in age_stats.index:
    print(f"\nKategori: {category}")
    print(f"Jumlah member: {age_stats.loc[category, ('id', 'count')]} orang")
    print(f"Rata-rata waktu: {age_stats.loc[category, ('avg_time_in_gym', 'mean')]} menit")
    print(f"Waktu minimum: {age_stats.loc[category, ('avg_time_in_gym', 'min')]} menit")
    print(f"Waktu maksimum: {age_stats.loc[category, ('avg_time_in_gym', 'max')]} menit")

# Statistik per kategori umur dan gender
print("\nStatistik per kategori umur dan gender:")
age_gender_stats = df.groupby(['age_category', 'gender'])['avg_time_in_gym'].agg(['count', 'mean', 'min', 'max']).round(2)
print(age_gender_stats)
```

```python
# 4. Mengurutkan dan menampilkan data
print("===== 4. Data Terurut Berdasarkan Waktu di Gym (Top 10) =====")
sorted_df = df.sort_values('avg_time_in_gym', ascending=False).head(10)
for idx, row in sorted_df.iterrows():
    print(f"\nPeringkat {sorted_df.index.get_loc(idx) + 1}:")
    print(f"ID: {row['id']}")
    print(f"Umur: {row['Age']} tahun")
    print(f"Gender: {row['gender']}")
    print(f"Waktu di Gym: {row['avg_time_in_gym']} menit")
print("\n")
```

```python
# 3. Menampilkan 10 data pertama menggunakan iterrows
print("===== 3. 10 Data Pertama =====")
for idx, row in df.head(10).iterrows():
    print(f"\nMember ke-{idx + 1}:")
    print(f"Umur: {row['Age']} tahun")
    print(f"Gender: {row['gender']}")
    print(f"Tipe Membership: {row['abonoment_type']}")
    print(f"Waktu Check-in: {row['avg_time_check_in']}")
    print(f"Waktu Check-out: {row['avg_time_check_out']}")
    print(f"Total Waktu di Gym: {row['avg_time_in_gym']} menit")
print("\n")
```

```python
# 2. Menampilkan deskripsi data
print("===== 2. Deskripsi Data =====")
print(df.describe())
print("\nInformasi kolom:")
print(df.info())
print("\n")
```
